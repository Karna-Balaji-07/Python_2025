```python
'''
1. Input a number and find the sum of first and last digit of the number
# input = 1234456
# output =  7  ( 1 + 6 )
'''

def solution1(n):
    sums = 0
    s = str(n)
    sums = int(s[0]) + int(s[-1])
    return sums

def solution2(n):
    last = n % 10
    while n >= 10:
        n //= 10
    first = n
    return last + first

n = 123456
print(solution1(n))
print(solution2(n))
```

```python
''' 
2. PRODUCT OF DIGITS
the task is to find the product of the digits
n = 12345
result = 120
'''

def solution1(n):
    num = n
    product = 1
    while num > 0:
        rem = num % 10
        product *= rem
        num //= 10

    return product

n = 12345
print(solution1(n))
```

```python
'''
3. PALINDROME NUMBER
The task is to check whether the number is palindrome or not
input = 121
output = True
input = -121
output = False
'''

def solution31(n):
    if n < 0:
        return False
    rev = 0
    num = n
    while num > 0:
        temp = num % 10
        rev = rev*10 + temp
        num //= 10
    if rev == n:
        return True
    return False

n1 = 121
n2 = -121
print(solution31(n1))
print(solution31(n2))
```

```python
''' 
4. THE POWER OF A NUMBER
The task is to find the power of a number using for loop
input = b = 2 ,e = 5
output = 32
'''

import math as m
def solution41(b,e):
    result = m.pow(b,e)
    return result

def solution42(b,e):
    return b ** e

def solution43(b,e):
    result = 1
    for i in range(e):
        result *= b
    return result


b = 2
e = 5
print(solution41(b,e))
print(solution42(b,e))
print(solution43(b,e))
```

```python
'''
5. FIBONACCI SERIES
The task is to find the fibonacci series till n
'''

def solution51(n):
    a = 0
    b = 1
    c = 0
    for i in range(1,n):
        a=b
        b=c
        c = a+b
        print(c,end=" ")

n = 10
solution51(n)
```

```python
''' 
6. FACTORS OF N
The task is to find all the factors of a number
'''

def solution61(n):
    result = []
    for i in range(1,n+1):
        if n % i == 0:
            result.append(i)
    return result

n = 24
print(solution61(n))
```

```python
''' 
7. PRIME NUMBER 
The task is to check whether the given number is prime or not
input = 17
output = prime
input  4
output Not prime
'''

def solution71(n):
    if n <= 1:
        return "Not prime"
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return "Not prime"
    return "Prime"

# Test
print(solution71(17))  # Prime
print(solution71(4))   # Not prime
```

```python
''' 
8. PRIME NUMBERS TILL N
The task is to find all the prime numbers till N
'''

def solution81(n):
    result = []
    for i in range(2, n+1):
        flag = 1
        for j in range(2, i//2+1):
            if i % j == 0:
                flag = 0
                break
                
        if flag == 1:
            result.append(i)
    return result
    

n = 20
print(solution81(n))
```

```python
''' 
9. PRIME FACTORS
Find all the prime factors of n
n = 10
output = 2,5

'''

def solution91(n):
    result = []
    for i in range(2,n+1):
        if n % i == 0:
          
            flag = 1
            for j in range(2,int(i**0.5)+1):
                if i % j == 0:
                    flag = 0
                    break
            if flag == 1:
                result.append(i)
    return result

n = 10
print(solution91(n))

```

```python
''' 
10. LCM 
The task is to find the least common multiples of 2 numbers
input = 5,10
output = 10
'''

def solution101(n1,n2):
    maxi = max(n1,n2)
    mini = min(n1,n2)
    for i in range(maxi, n1*n2+1, maxi):
        if i % mini == 0:
            return i
        
    return -1

print(solution101(8,4))
```

```python
''' 
11. HCF of two numbers
the task is to find the greatest common divisor of two numbers
'''

def solution1101(a,b):
    gcd = 0
    for i in range(1, min(a,b)+1):
        if a % i == 0 and b % i == 0:
            gcd = i
    return gcd

def solution1102(a,b):
    gcd = 0
    for i in range(min(a,b),0,-1):
        if a % i == 0 and b % i == 0:
            return i
    return -1


def solution1103(a,b):
    while a >0 and b > 0:
        if a > b:
            a = a % b
        else:
            b = b % a
    if a == 0:
        return b
    return a

print(solution1101(15, 30))
print(solution1102(15,30))
print(solution1103(15,30))
```

```python
''' 
12. PERFECT NUMBER
The task is to check whether the given number is prefect or not
'''

def solution1201(n):
    sums = 0
    for i in range(1,n):
        if n % i == 0:
            sums += i
    return sums == n

n = 28
print(solution1201(n))
```

```python
''' 
13. STRONG NUMBERS
The task is to check whether the given number is a strong number or not
'''

def fact(n):
    if n == 1:
        return 1
    return n * fact(n-1)

def solution1301(n):
    sums = 0
    num = n
    while num > 0:
        temp = num % 10
        res = fact(temp)
        sums += res
        num //= 10
    
    return sums == n

n = 145111
print(solution1301(n))
```

```python
''' 
14. SQUARE PATTERN
'''

def solution14(n):
    for i in range(n):
        for j in range(n):
            print('*',end="")
        print()

solution14(5)
```

```python
'''
15. HOLLOW SQUARE PATTERN
'''
def solution15(n):
    for i in range(n):
        for j in range(n):
            if i == 0 or i == n-1:
                print("*",end="")
            else:
                if j ==0 or j == n-1:
                    print("*",end="")
                else:
                    print(" ",end="")
        print()

solution15(5)
```

```python
''' 
31. Print chessboard number pattern with 1 and 0
10101
01010
10101
01010
10101
'''
def solution31(row, col):
    for i in range(row):
        for j in range(col):
            if i % 2 == 0:
                if j % 2 == 1:
                    print("0",end="")
                else:
                    print("1",end="")
            else:
                if j % 2 == 0:
                    print("0",end="")
                else:
                    print("1",end="")
        print()
solution31(5,5)
```

```python
''' 
32. Print box number pattern with plus in center
11011
11011
00000
11011
11011
'''

def solution32(row, col):
    midrow = row//2
    midcol = col // 2
    for i in range(row):
        for j in range(col):
            if i == midrow or j == midcol:
                print("0",end="")
            else:
                print("1",end="")
        print()

solution32(5,5)           
```

```python
''' 
33.  Print box number pattern of 1 and 0 with cross center
'''
def solution33(row, col):
    for i in range(1,row+1):
        for j in range(1,col+1):
            if i == j or j == (col+1)-i:
                print("1",end="")
            else:
                print("0",end="")
        print()
solution33(5,5)
```
