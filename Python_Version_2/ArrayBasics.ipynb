```python
''' 
1. Print value and index of the given array 
'''
def solution1(arr):
    for key, value in enumerate(arr):
        print(f'Key : {key} >> Value : {value}')
arr = [12,43,675,896,243]
solution1(arr)
```

```python
''' 
2. Reverse an array
'''
def solution2(arr):
    for i in range(len(arr)-1,-1,-1):
        print(arr[i],end=" ")

def solution200(arr):
    first = 0
    last = len(arr)-1
    while first < last:
        arr[first],arr[last] = arr[last],arr[first]
        first += 1
        last -=1 
    return arr

def solution201(arr):
    n = len(arr)
    for i in range(n//2):
        temp = arr[i]
        arr[i] = arr[n-i-1]
        arr[n-i-1] = temp
    return arr

    
arr = [1,2,3,4,5,6]
print(solution201(arr))
```

```python
''' 
3. Print alternate elements of an array
'''
def solution3(arr):
    for i in range(0,len(arr),2):
        print(arr[i],end=" ")
arr = [1, 2, 9, 5, 6, 7, 8, 2, 3]
solution3(arr)
    
```

```python
''' 
4. Seperate odd and even element of the array
'''

def solution4a(arr):
    odd = []
    even = []
    for i in arr:
        if i % 2 == 0:
            even.append(i)
        else:
            odd.append(i)
    return odd,even
    
arr = [1,2,45,7,4,8,4,8,5,7,34,7]
print(solution4a(arr))
```

```python
''' 
5. Find the number of times the target appears in the array
'''

def solution5a(arr, target):
    count = 0
    for i in arr:
        if i == target:
            count += 1
    return count

arr = [1,2,2,3,4,5,2,2,5,87,2,76,8,2]
print(solution5a(arr,2))
```

```python
'''
6. Check if the array is sorted forward, backward, no not at all
'''

def solution6a(arr):
    maxi = max(arr)
    for i in range(1, len(arr)):
        
        if arr[0] == maxi:
            if arr[i] > arr[i-1]:
                return False
        else:
            if arr[i] < arr[i-1]:
                return False
    return True

def solution6b(arr):
    ascending = True
    descending = True
    for i in range(len(arr)-1):
        if arr[i] > arr[i+1]:
            ascending = False
        elif arr[i] < arr[i+1]:
            descending = False
    if ascending:
        print("Sorted in ascending order")
    elif descending:
        print("Sorted in descending order")
    else:
        print("Not sorted")

arr = [7,6,5,4,3,2,1]
arr = [1,2,3,4,5,6,7]
arr = [1,3,5,6,2.9,0,1]
solution6b(arr)
```

```python
''' 
7. Count duplicates and unique elements in the array
'''
def solution7a(arr):
    dicts = {}
    for i in arr:
        dicts[i] = dicts.get(i,0)+1
    duplicates = 0
    unique = 0
    for values in dicts.values():
        if values == 1:
            unique += 1
        else:
            duplicates += 1
    return unique, duplicates

arr = [1,2,2,3,4,4,4,5,6]
print(solution7a(arr))
```
