# Traversal Of Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def traversal(head):
    curr = head
    while curr is not None:
        print(curr.data, end=" ")
        curr = curr.next
    print()

node = Node(10)
node.next = Node(20)
node.next.next = Node(30)
node.next.next.next = Node(40)
node.next.next.next.next = Node(50)

traversal(node)
```

# Searching in a Linked List

```python
def search(head, value):
    curr = head
    while curr is not None:
        if curr.data == value:
            return True
        curr = curr.next
    return False

node = Node(10)
node.next = Node(20)
node.next.next = Node(30)
node.next.next.next = Node(40)
node.next.next.next.next = Node(50)

result = search(node, 20)
results = search(node, 33)
print(result)
print(results)
```

# Length of the Linked List

```python
def length(head):
    count = 0
    curr = head
    while curr is not None:
        count += 1
        curr = curr.next

    return count

result = length(node)
print(result)
```

# Insert a Node at the front of the Linked List

```python
def insert_front(head, value):
    newnode = Node(value)
    newnode.next = head
    return newnode # returns the new node as the head of the linked list

result = insert_front(node,5 )
traversal(result)
```

# Insert a node at the end of the linked list

```python
def insert_end(head, value):
    newNode = Node(value)
    
    curr = head
    if head is None:
        return newNode
    while curr.next is not None:
        curr = curr.next
    
    curr.next = newNode
    return head

result = insert_end(node, 60)
traversal(result) 
```

# Insert a Node at a given position in the Linked List 

```python
def insert_position(head, pos, value):
    if pos < 1:
        return head
    if pos == 1:
        newnode = Node(value)
        newnode.next = head
        return newnode
    
    curr = head
    for _ in range(1, pos-1):
        if curr == None:
            break
        curr = curr.next

    if curr is None:
        return head
    newnode = Node(value)
    newnode.next = curr.next
    curr.next = newnode
    return head

result = insert_position(node, 2, 25)
traversal(result)
```

# Removal of First node of the linked list

```python
def removal_first(head):
    temp = head
    head = head.next
    del temp
    return head

result = removal_first(node)
traversal(result)
```

# Removal of Last Node of the linked list

```python
def removal_last(head):
    if not head:
        return None
    if not head.next:
        return None
    
    curr = head
    while curr.next.next is not None:
        curr = curr.next

    curr.next = None
    return head

result = removal_last(node)
traversal(result)
```

# Removal of a node at a specified position in the linked list

```python
def removal_pos(head, pos):
    temp = head
    prev = None
    if temp is None:
        return head
    
    if pos == 1:
        head = temp.next
        return head
    
    for _ in range(1, pos):
        prev = temp
        temp = temp.next
        if temp is None:
            return head
        
    if temp is not None:
        prev.next = temp.next

    return head

result = removal_pos(node, 3)
traversal(result)
```

# Reverse of Linked List

```python
def approach1(head):
    curr = head
    prev = None
    while curr is not None:
        nextnode = curr.next
        curr.next = prev
        prev = curr
        curr = nextnode
    return prev

def approach2(head):
    stack = []
    temp = head
    while temp.next is not None:
        stack.append(temp.data)
        temp = temp.next

    head = temp
    while stack:
        temp.next = stack.pop()
        temp = temp.next

    temp.next = None
    return head

def traversal(head):
    temp = head
    while temp is not None:
        print(temp.data,end=" ")
        temp = temp.next

    print()
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)

result2 = approach1(head)

traversal(result2)
```
