```python
''' 
1. Second largest
Given an array of positive integers arr[], return the second largest element from the array. 
If the second largest element doesn't exist then return -1.

Note: The second largest element should not be equal to the largest element.
'''

def solution1(arr):
    s = list(set(arr))
    s.sort()
    if len(s) != 1:
        return s[-2]
    else:
        return -1


def solution2(arr):
    maxi1 = float('-inf')
    for i in arr:
        if i > maxi1:
            maxi1 = i
    maxi2 = float('inf')
    for i in arr:
        if i > maxi2 and i != maxi1:
            maxi2 = i
    return maxi2

def solution3(arr):
    maxi = -1
    second = -1
    for i in range(len(arr)):
        if arr[i] > maxi:
            second = maxi
            maxi = arr[i]
        elif arr[i] < maxi and second < arr[i]:
            second = arr[i]
        
    return second
            


arr = [12, 35, 1, 10, 34, 1]
print(solution3(arr))
```

```python
''' 
2. Move all zeroes to the end
You are given an array arr[] of non-negative integers.
 Your task is to move all the zeros in the array to the right end while maintaining the relative order of the non-zero elements.
 The operation must be performed in place, meaning you should not use extra space for another array
'''

def solution1(arr):
    for i in range(1, len(arr)):
        if arr[i-1] == 0:
            arr[i],arr[i-1] = arr[i-1],arr[i]
        
    return arr

arr = [1,2,0,4,3,0,5,0]
print(solution1(arr))
```
