```python
''' 
1. Find the second smallest and second largest element in the array
'''

def solution1(arr):
    arrs = list(set(arr))
    arrs.sort()
    return arrs[1],arrs[-2]

def solution2(arr):
    mini1 = float('inf')
    mini2 = float('inf')
    maxi1 = float('-inf')
    maxi2 = float('-inf')
    for i in arr:
        if i < mini1:
            mini1 = i
    for i in arr:
        if i < mini2 and i != mini1:
            mini2 = i
    
    for i in arr:
        if i > maxi1:
            maxi1 = i
    for i in arr:
        if i > maxi2 and i != maxi1:
            maxi2 = i
    if mini2 == float('inf') and maxi2 == float('-inf'):
        return -1
    else:
        return mini2, maxi2

def solution3(arr):
    small = float('inf')
    second_small = float('inf')
    large  = float('-inf')
    second_large = float('-inf')

    for i in arr:
        if i < small:
            second_small = small
            small = i
        elif i < second_small and i != small:
            second_small = i

        elif i > large:
            second_large = large
            large = i
        elif i > second_large and i != large:
            second_large = i

    return second_small,second_large

arr = [1,2,4,7,7,5]
print(solution3(arr))
```

```python
''' 
2. check if the array is sorted or not
'''

def solution1(arr):
    arrs = []
    arrs[:] = arr[:]
    arrs.sort()
    return arrs == arr

def solution2(arr):
    asc = True
    desc = True
    for i in range(len(arr)-1):
        if arr[i] > arr[i+1]:
            asc = False
        if arr[i] < arr[i+1]:
            desc = False
        
    if asc:
        return "ASC sorted"
    elif desc:
        return "DESC sorted"
    else:
        return False

arr = [5,4,3,7,8,9]
print(solution2(arr))
```

```python
''' 
Left rotate an array by 1  place
'''

def solution1(arr):
    arr[1:],arr[:1] = arr[:1],arr[1:]
    return arr

arr = [1,2,3,4,5]
print(solution1(arr))
```
