# BINARY SEARCH

1. Binary Search algorithm

```python
def binary(arr, target):
    first = 0
    last = len(arr)-1
    while first <= last:
        mid = (first+last) //2
        if arr[mid] == target:
            return mid
        elif arr[mid] > target:
            last = mid -1
        else:
            first + mid + 1
        
    return -1

def binary1(arr, target, first, last):
    if first <= last:
        mid = (first+ last) //2
        if arr[mid] == target:
            return mid
        if arr[mid] > target:
            return binary1(arr, target, first, mid-1)
        else:
            return binary1(arr, target, mid+1, last)
    return -1


arr = [2,4,5,7,8,11,22,24,27]
target = 8
print(binary(arr, target))
print(binary1(arr, target, 0, len(arr)))
```

2. Find first and last position of element in sorted array 

```python
def first_element(arr, target):
    first = 0
    last = len(arr)-1
    first_occ = -1
    while first <= last:
        mid = (first+last) // 2
        if arr[mid] == target:
            last = mid-1
            first_occ = mid
        elif arr[mid] > target:
            last = mid -1
        else:
            first = mid+1
    return first_occ


def last_element(arr, target):
    first = 0
    last = len(arr)-1
    last_occ = -1
    while first <= last:
        mid = (first+last) // 2
        if arr[mid] == target:
            first = mid+1
            last_occ = mid

        elif arr[mid] > target:
            last = mid-1
        else:
            first = mid+1
    return last_occ

arr = [2,3,5,8,8,8,9,66]
target = 8
print(first_element(arr, target))
print(last_element(arr, target))
```

3. Unique number of occurences

```python
def unique(arr):
    dicts = {}
    for i in arr:
        dicts[i] = dicts.get(i,0)+1
    arrs = list(dicts.values())
    return len(arrs) == len(set(arrs))

arr = [1,2,2,1,1,3]
print(unique(arr))
```

4. Square root of a number using binary search

```python
def square(n):
    first = 0
    last = n
    result = 0
    while first <= last:
        mid = (first+last)//2
        if mid * mid == n:
            return mid
        elif mid*mid < n:
            result = mid
            first += 1
        else:
            last -=1

    return result

def square1(n):
    first = 0
    last = n
    while first <= last:
        mid = (first+last) // 2
        val = mid * mid
        if val <= n:
            first = mid+1
        else:
            last = mid-1

    return last

n = 9
print(square1(n))
```

5. Kth smallest element in sorted matrix

```python
def sorted(matrix, k):
    count = 1
    for arr in matrix:
        for i in arr:
            if count == k:
                return i
            count += 1
    return -1


matrix = [[1,5,9],[10,11,13],[12,13,15]]
k = 8
print(sorted(matrix,k))
```

6. Median of two sorted arrays

```python
def median(nums1, nums2):
    nums1.extend(nums2)
    nums1.sort()
    if len(nums1) % 2 != 0:
        mid = len(nums1) //2
        return nums1[mid]
    else:
        mid = len(nums1) // 2
        first = nums1[mid]
        last = nums1[mid-1]
        result = float((first+last)/2)
        return float(result)
    
nums1 = [1,2]
nums2 = [3,4]
print(median(nums1, nums2))
```

4. Remove duplicates from sorted array

```python
def duplicates(arr):
    i = 0
    for j in range(1, len(arr)):
        if arr[i] != arr[j]:
            i += 1
            arr[i] = arr[j]
    
    return arr[:i+1]

arr = [0,0,1,1,1,2,2,3,3,4]
print(duplicates(arr))
```
